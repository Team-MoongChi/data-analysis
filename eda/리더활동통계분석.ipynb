# 최근 6개월간 리더 수 분석


```python
# 날짜 형식 변환
participants_df['joined_at'] = pd.to_datetime(participants_df['joined_at'])
user_df['created_at'] = pd.to_datetime(user_df['created_at'])
user_df['updated_at'] = pd.to_datetime(user_df['updated_at'])
group_boards_df['created_at'] = pd.to_datetime(group_boards_df['created_at'])
group_boards_df['updated_at'] = pd.to_datetime(group_boards_df['updated_at'])
group_boards_df['deadline'] = pd.to_datetime(group_boards_df['deadline'])

# 기준 날짜 설정 (2025-06-02)
reference_date = pd.to_datetime('2025-06-02')

# 6개월 전 날짜 계산
six_months_ago = reference_date - pd.DateOffset(months=6)

# 최근 6개월간 활동한 리더 필터링
recent_leaders = participants_df[(participants_df['role'] == '리더') &
                                 (participants_df['joined_at'] >= six_months_ago)]
monthly_leaders = recent_leaders.groupby(pd.Grouper(key='joined_at', freq='ME')).size().reset_index()
monthly_leaders.columns = ['month', 'leader_count']

# 월별 리더 수 시각화
plt.figure(figsize=(10, 6))
plt.bar(monthly_leaders['month'].dt.strftime('%Y-%m'), monthly_leaders['leader_count'])
plt.title('최근 6개월 월별 리더 수 추이', fontsize=15)
plt.xlabel('월', fontsize=12)
plt.ylabel('리더 수', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 총 리더 수
total_leaders = recent_leaders['user_id'].nunique()
print(f"최근 6개월간 총 활동 리더 수: {total_leaders}명")

# 최근 6개월 추이 구하기
print("월별 리더 수 추이:")
for _, row in monthly_leaders.iterrows():
  print(f"{row['month'].strftime('%Y-%m')}: {row['leader_count']}명")

plt.show()
```

# 리더 1회 이상 재참여율


```python
# 리더로 참여한 횟수가 2회 이상인 사용자 비율 계산
leader_participation = participants_df[participants_df['role'] == '리더'].groupby('user_id').size().reset_index()
leader_participation.columns = ['user_id', 'participation_count']

# 총 리더 수
total_unique_leaders = leader_participation['user_id'].nunique()

# 2회 이상 참여한 리더 수
repeated_leaders = leader_participation[leader_participation['participation_count'] >= 2]['user_id'].nunique()

# 재참여율 계산
repeated_rate = (repeated_leaders / total_unique_leaders) * 100

# 시각화
plt.figure(figsize=(10, 6))
participation_counts = leader_participation['participation_count'].value_counts().sort_index()

# 그래프 시각화
plt.bar(participation_counts.index, participation_counts.values)
plt.title('리더 참여 횟수 분포', fontsize=15)
plt.xlabel('참여 횟수', fontsize=12)
plt.ylabel('리더 수', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 1회 이상 재참여율 텍스트로 표시
plt.text(0.5, 0.9, f'1회 이상 재참여율: {repeated_rate:.1f}%',
         transform=plt.gca().transAxes, fontsize=13,
         bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.5'))

print(f"총 리더 수: {total_unique_leaders}명")
print(f"2회 이상 참여한 리더 수: {repeated_leaders}명")
print(f"1회 이상 재참여율: {repeated_rate:.2f}%")

plt.tight_layout()
plt.show()
```

# 리더 평균 공구 개설 수


```python
leader_avg_boards = leader_participation['participation_count'].mean()

# 리더별 개설 수 분포 히스토그램
plt.figure(figsize=(10, 6))
sns.histplot(leader_participation['participation_count'], bins=20, kde=True)
plt.axvline(x=leader_avg_boards, color='red', linestyle='--', label=f'평균: {leader_avg_boards:.2f}')
plt.title('리더별 공구 개설 수 분포', fontsize=15)
plt.xlabel('공구 개설 수', fontsize=12)
plt.ylabel('리더 수', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

print(f"리더 1인당 평균 개설 수: {leader_avg_boards:.2f}회")
print(f"최소 개설 수: {leader_participation['participation_count'].min()}회")
print(f"최대 개설 수: {leader_participation['participation_count'].max()}회")
print(f"중앙값 개설 수: {leader_participation['participation_count'].median()}회")

plt.tight_layout()
plt.show()
```

# 활동이 많은 리더 상위 10명 순위

- 개설 수 기준
- 매너온도 표시


```python
top_leaders = leader_participation.sort_values('participation_count', ascending=False).head(10)

# 상위 리더들의 정보 추가 - 사용자 정보와 결합
top_leaders_info = pd.merge(top_leaders, user_df[['user_id', 'name', 'manner_leader', 'address']], on='user_id')

fig, ax = plt.subplots(figsize=(12, 6))

# 매너 점수를 정규화해 색상으로 사용
cmap = plt.colormaps.get_cmap('RdYlGn')
manner_normalized = (top_leaders_info['manner_leader'] - top_leaders_info['manner_leader'].min()) / \
    (top_leaders_info['manner_leader'].max() - top_leaders_info['manner_leader'].min())

# 수평 막대그래프
bars = ax.barh(top_leaders_info['name'], top_leaders_info['participation_count'],
               color=[cmap(score) for score in manner_normalized])

# 제목 및 축 설정
ax.set_title('공구 활동이 많은 상위 10명 리더', fontsize=15)
ax.set_xlabel('공구 개설 수', fontsize=12)
ax.set_ylabel('리더 닉네임', fontsize=12)
ax.grid(axis='x', linestyle='--', alpha=0.7)

# 컬러바 추가
norm = plt.Normalize(vmin=top_leaders_info['manner_leader'].min(),
                     vmax=top_leaders_info['manner_leader'].max())
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax)
cbar.set_label('리더 매너 점수', fontsize=10, family='NanumGothic')

# 정보 출력
print("공구 활동이 많은 상위 10명 리더:")
for i, row in top_leaders_info.iterrows():
  print(f"{row['name']} - {row['participation_count']}회 개설 (매너점수: {row['manner_leader']:.2f}, 지역: {row['address'].split()[0]})")

plt.tight_layout()
plt.show()
```

# 지역별 리더 분포


```python
# user_df에서 지역 정보 추출 (도/시 단위)
user_df['region'] = user_df['address'].apply(lambda x: x.split()[0] if ' ' in x else x)

# 리더로 활동한 사용자 ID 목록
leader_ids = participants_df[participants_df['role'] == '리더']['user_id'].unique()

# 리더로 활동한 사용자들의 지역 정보
leaders_by_region = user_df[user_df['user_id'].isin(leader_ids)].groupby('region').size().reset_index()
leaders_by_region.columns = ['region', 'leader_count']
leaders_by_region = leaders_by_region.sort_values('leader_count', ascending=False)

# 상위 10개 지역 시각화
plt.figure(figsize=(12, 7))
sns.barplot(x='leader_count', y='region', data=leaders_by_region.head(10), palette='viridis')
plt.title('지역별 리더 분포 (상위 10개 지역)', fontsize=15)
plt.xlabel('리더 수', fontsize=12)
plt.ylabel('지역', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

print("지역별 리더 분포 (상위 5개 지역):")
for i, row in leaders_by_region.head(5).iterrows():
  print(f"{row['region']}: {row['leader_count']}명 ({row['leader_count'] / len(leader_ids) * 100:.1f}%)")

print("\n리더 활동이 적은 지역 (하위 5개 지역):")
for i, row in leaders_by_region.tail(5).iterrows():
  print(f"{row['region']}: {row['leader_count']}명 ({row['leader_count'] / len(leader_ids) * 100:.1f}%)")

plt.tight_layout()
plt.show()

# 지도 시각화를 위한 지역별 리더 분포 데이터 준비
# 여기서는 지역 정보만 시각화
```

# 리더별 공구 개설 수(활동성)와 매너 점수 관계 분석


```python
leader_activity_satisfaction = pd.merge(
    leader_participation,
    user_df[['user_id', 'manner_leader', 'manner_participant']],
    on='user_id'
)

# 산점도 시각화
plt.figure(figsize=(10, 8))
scatter = plt.scatter(leader_activity_satisfaction['participation_count'],
                      leader_activity_satisfaction['manner_leader'],
                      c=leader_activity_satisfaction['manner_participant'],
                      cmap='viridis',
                      alpha=0.7,
                      s=80)

plt.title('리더 활동 수와 매너 점수의 관계', fontsize=15)
plt.xlabel('리더 활동 수', fontsize=12)
plt.ylabel('리더 매너 점수', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# 컬러바 추가
cbar = plt.colorbar(scatter)
cbar.set_label('참여자 매너 점수', fontsize=10)

# 상관관계 계산
corr_activity_manner = np.corrcoef(leader_activity_satisfaction['participation_count'],
                                   leader_activity_satisfaction['manner_leader'])[0, 1]

# 평균 추세선 추가
z = np.polyfit(leader_activity_satisfaction['participation_count'],
               leader_activity_satisfaction['manner_leader'], 1)
p = np.poly1d(z)
plt.plot(sorted(leader_activity_satisfaction['participation_count'].unique()),
         p(sorted(leader_activity_satisfaction['participation_count'].unique())),
         "r--", linewidth=2)

# 상관관계 정보 추가
plt.text(0.05, 0.95, f'상관계수: {corr_activity_manner:.3f}',
         transform=plt.gca().transAxes, fontsize=12,
         bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.5'))

print(f"리더 활동 수와 리더 매너 점수의 상관계수: {corr_activity_manner:.3f}")

plt.tight_layout()
plt.show()
```

# 카테고리별 공구 활동 분포


```python
# 카테고리 대분류 정보 추출
large_categories = category_df[category_df['level'] == 'large'][['category_id', 'name']]
large_categories.columns = ['parent_category_id', 'large_category_name']

# 중분류 카테고리에 대분류 매핑
medium_categories = category_df[category_df['level'] == 'medium'].copy()
medium_categories = pd.merge(medium_categories, large_categories, on='parent_category_id', how='left')

# 공구 게시판과 공구 상품 정보 결합
group_product_info = pd.merge(group_boards_df, group_product_df,
                              left_on='group_product_id', right_on='group_product_id', how='left')

# 카테고리 정보 추가
group_product_info = pd.merge(group_product_info, category_df[['category_id', 'name']],
                              left_on='category_id', right_on='category_id', how='left')
group_product_info.rename(columns={'name': 'category_name'}, inplace=True)

# 카테고리별 공구 수 집계
category_distribution = group_product_info.groupby('category_name').size().reset_index()
category_distribution.columns = ['category_name', 'board_count']
category_distribution = category_distribution.sort_values('board_count', ascending=False)

# 상위 10개 카테고리 시각화
plt.figure(figsize=(12, 7))
top_categories = category_distribution.head(10)
sns.barplot(x='board_count', y='category_name', data=top_categories, palette='muted')
plt.title('카테고리별 공구 게시판 수 (상위 10개)', fontsize=15)
plt.xlabel('공구 게시판 수', fontsize=12)
plt.ylabel('카테고리', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

print("카테고리별 공구 게시판 수 (상위 5개):")
for i, row in top_categories.head(5).iterrows():
  print(f"{row['category_name']}: {row['board_count']}개")

plt.tight_layout()
plt.show()
```

# 상위 10개 카테고리와 관련된 리더 활동 분석


```python
# 상위 카테고리 ID 추출
top_category_ids = group_product_info[group_product_info['category_name'].isin(top_categories['category_name'])]['category_id'].unique()

# 상위 카테고리의 리더 활동 데이터
top_category_leaders = group_product_info[group_product_info['category_id'].isin(top_category_ids)]
top_category_leaders = pd.merge(
    top_category_leaders,
    participants_df[participants_df['role'] == '리더'][['group_board_id', 'user_id']],
    on='group_board_id',
    how='left'
)

# 리더 정보 추가
top_category_leaders = pd.merge(
    top_category_leaders,
    user_df[['user_id', 'manner_leader', 'region']],
    on='user_id',
    how='left'
)

# 카테고리별 평균 리더 매너 점수
category_manner_scores = top_category_leaders.groupby('category_name')['manner_leader'].mean().reset_index()
category_manner_scores = category_manner_scores.sort_values('manner_leader', ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(x='manner_leader', y='category_name', data=category_manner_scores, palette='mako')
plt.title('카테고리별 리더 평균 매너 점수', fontsize=15)
plt.xlabel('평균 매너 점수', fontsize=12)
plt.ylabel('카테고리', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

print("카테고리별 평균 리더 매너 점수:")
for i, row in category_manner_scores.iterrows():
  print(f"{row['category_name']}: {row['manner_leader']:.2f}")

plt.tight_layout()
plt.show()
```
